```python
import numpy as np
q1 = np.quantile(df['Annual Income (k$)'],0.25)
print('q1',q1)

q2 = np.quantile(df['Annual Income (k$)'],0.5)
print('q2',q2)

q3 = np.quantile(df['Annual Income (k$)'],0.75)
print('q3',q3)

iqr = (q3 - q1)
print('iqr',iqr)
min,max = q3-1.5*iqr,q1+1.5*iqr
iqrrange = [min,max]
iqrrange
```

```python
range = df[(df['Annual Income (k$)'] >= min) & (df['Annual Income (k$)'] <= max)]
range.count()
```
